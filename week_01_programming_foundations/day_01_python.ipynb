```python
age = 37

def age_check(age):
    if age < 18:
        age_group = "minor"
    elif age < 30:
        age_group = "young adult"
    else:
        age_group = "adult"
    return age_group

age_group = age_check(age)
print(f"{name} is classified as an {age_group}.")
```

jobs = ["AI Engineer", "Product Manager", "Backend Developer", "Data Entry", "DevOps Engineer"]

def filter_jobs(jobs):
    suitable_jobs = []
    for job in jobs:
        if "Engineer" in job or "Developer" in job:
            suitable_jobs.append(job)
        else:
            continue
    return suitable_jobs

suitable_jobs = filter_jobs(jobs)
print(f"Suitable job roles based on my interests: {suitable_jobs}")

```python
jobs = ["AI Engineer", "Product Manager", "Backend Developer", "Data Entry", "DevOps Engineer"]

suitable_jobs = [job for job in jobs if "Engineer" in job or "Developer" in job]
# ^ List comprehension to filter suitable jobs
print(f"Suitable job roles based on my interests: {suitable_jobs}")
```

```python
# Career tracker challenge
current_skill = 1
weeks = 8

while current_skill <= weeks:
    print(f"Week {current_skill}: Skill level {current_skill}/8")
    current_skill += 1

print("Congratulations! You have completed the career tracker challenge.")
```
